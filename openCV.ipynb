```python
#croping
img_crop=img_resize[100:200,100:250,:]
```

```python
# to save the croped image to be used in future
cv2.imwrite('cropped_fruit.png',img_crop)
```

```python
img2=np.zeros((256,256,3))
```

```python
#drawing shapes on image
# top left point, bottom right point ,
cv2.rectangle(img2,pt1=(100,100),pt2=(200,200),color=(255,125,0),thickness=3)# to fill tickness=-1
```

```python
cv2_imshow(img2)
```

```python
cv2.circle(img2, center=(50, 50),radius=50, color=(0, 125, 225), thickness=-1)
```

```python
cv2_imshow(img2)
```

```python
org = (50, 100)  # Coordinates of the text
font = cv2.FONT_HERSHEY_SIMPLEX
font_scale = 2
font_thickness = 4
font_color = (255, 255, 255)  # BGR color format

# Put text on the image
cv2.putText(img2, text='HIIIIII', org=org, fontFace=font, fontScale=font_scale,
            color=font_color, thickness=font_thickness)
```

```python
cv2_imshow(img2)
```
